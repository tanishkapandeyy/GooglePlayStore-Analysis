Task 5 – Rating and Review Analysis

## Objective

To compare average ratings and total review counts across the top 10 app categories by installs.

## Dataset

Google Play Store Dataset

## Transformations Applied

* Filtered categories with average rating ≥ 4.0
* Filtered app size > 10 MB
* Selected apps updated during January
* Identified top 10 categories based on installs

## KPI Measured

* Average Rating
* Total Reviews

## Visualization

Grouped Bar Chart comparing ratings and review volumes across categories.

## Time Restriction

The chart is available only between 3 PM IST and 5 PM IST.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")
```

```python
apps_df = pd.read_csv(
    r"C:\Users\HP\Downloads\archive\googleplaystore.csv"
)

apps_df.head()
```

```python
apps_df.columns.tolist()
```

```python
apps_df = apps_df[apps_df['Size'] != 'Varies with device']

apps_df['Size'] = apps_df['Size'].str.replace('M', '')

apps_df['Size'] = apps_df['Size'].str.replace('k', '')

apps_df['Size'] = pd.to_numeric(
    apps_df['Size'],
    errors='coerce'
)

apps_df.head()
```

```python
apps_df['Installs'] = (
    apps_df['Installs']
    .str.replace(',', '')
    .str.replace('+', '')
)

apps_df['Installs'] = pd.to_numeric(
    apps_df['Installs'],
    errors='coerce'
)
```

```python
apps_df['Reviews'] = pd.to_numeric(
    apps_df['Reviews'],
    errors='coerce'
)
```

```python
apps_df['Last Updated'] = pd.to_datetime(
    apps_df['Last Updated'],
    errors='coerce'
)

apps_df['Month'] = apps_df['Last Updated'].dt.month

apps_df[['Last Updated', 'Month']].head()
```

```python
filtered_df = apps_df[
    (apps_df['Rating'] >= 4.0) &
    (apps_df['Size'] > 10) &
    (apps_df['Month'] == 1)
]

filtered_df.shape
```

```python
top_categories = (
    filtered_df
    .groupby('Category')['Installs']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_categories
```

```python
chart_df = (
    filtered_df[
        filtered_df['Category'].isin(top_categories.index)
    ]
    .groupby('Category')
    .agg({
        'Rating':'mean',
        'Reviews':'sum'
    })
    .reset_index()
)

chart_df
```

```python
chart_df['Reviews_Scaled'] = chart_df['Reviews'] / 100000
```

```python
from datetime import datetime
import pytz

ist = pytz.timezone('Asia/Kolkata')
current_time = datetime.now(ist)

if 15 <= current_time.hour < 17:

    import numpy as np
    import matplotlib.pyplot as plt

    chart_df['Reviews_Scaled'] = chart_df['Reviews'] / 100000

    x = np.arange(len(chart_df))
    width = 0.4

    plt.figure(figsize=(14,6))

    plt.bar(
        x - width/2,
        chart_df['Rating'],
        width,
        label='Average Rating'
    )

    plt.bar(
        x + width/2,
        chart_df['Reviews_Scaled'],
        width,
        label='Reviews (x100,000)'
    )

    plt.xticks(
        x,
        chart_df['Category'],
        rotation=45
    )

    plt.xlabel('Category')
    plt.ylabel('Value')

    plt.title(
        'Average Rating vs Total Reviews by Top 10 Categories'
    )

    plt.legend()
    plt.grid(axis='y', alpha=0.3)

    plt.tight_layout()
    plt.show()

else:
    print(
        "Grouped bar chart is available only between 3 PM IST and 5 PM IST."
    )
```
